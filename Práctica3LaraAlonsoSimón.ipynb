1 Escribe un programa analice el fichero mbox-short.txt e imprima por pantalla todas las ocurrencias de la expresión regular que escriba el usuario junto con el número de veces que aparece en el texto.

El programa debe tratar las posibles excepciones de tal manera que nunca se cierre a causa de un error de ejecución.

```
import re

# Abro el archivo
manejador = open('mbox-short.txt', 'r')
# Guardo el archivo leído en la variable texto
texto = manejador.read()

print('Escriba, por favor, una expresión regular:')
# Guardo la regex que escriba el usuario en una variable
regex_usuario = input()
try:
  # Encuentro todas las ocurrencias de la regex del usuario
  # Esta línea puede darme error si la regex introducida es inválida
  lista_ocurrencias = re.findall(regex_usuario, texto)
  # Guardo en una variable la longitud de la lista con las ocurrencias
  numero_ocurrencias = len(lista_ocurrencias)
  # Informo al usuario de cuántas veces aparece la regex en el texto
  print('La expresión regular', regex_usuario, 'aparece en el texto', numero_ocurrencias, 'veces.')
  # Si la regex aparece alguna vez en el texto, imprimo todas sus ocurrencias
  if len(lista_ocurrencias) > 0:
    # Aquí imprimo las ocurrencias en una columna
    print('Se muestran las ocurrencias en una columna:')
    # Recorro la lista e imprimo cada elemento
    for o in lista_ocurrencias:
      print(o)
    print('Se muestran las ocurrencias en una línea separadas por comas:')
    # Aquí imprimo todas las ocurrencias en una línea separadas por comas
    # Para eso hago un join en la lista de ocurrencias y le pongo el separador coma
    print(', '.join(lista_ocurrencias))
except:
  # Si la línea de arriba da error, se imprime lo siguiente
  print('La regex que introducido', regex_usuario, 'es una regex inválida.')
  print('Vuelva a intentarlo, por favor.')
```

2 Escribe un programa que analice el fichero mbox.txt e imprima por pantalla las direcciones de correo que han enviado correos más tarde de las 14:30.
Deben analizarse las líneas que presentan el siguiente formato, de tal manera que se devuelva el correo rjlowe@iupui.edu pero no el correo stephen.marquard@uct.ac.za:

Author: stephen.marquard@uct.ac.za

Date: 2008-01-05 09:12:07 -0500 (Sat, 05 Jan 2008)

[...]

Author: rjlowe@iupui.edu

Date: 2008-01-04 14:48:37 -0500 (Fri, 04 Jan 2008)

Para obtener la máxima calificación el programa debe incluir una función calcular_tarde que tome como argumento una fecha y devuelva verdadero si la fecha es posterior a las 14:30 y falso en caso contrario.

```
import re

# Abro el archivo
manejador = open('mbox.txt', 'r')
# Guardo el archivo leído en la variable texto
texto = manejador.read()

# Esta es la regex para sacar las dos líneas juntas que necesito
regex_contiguas = 'Author: .+\n.+'
# Esta es la lista de las dos líneas juntas que necesito
lista_contiguas = re.findall(regex_contiguas, texto)

# Función productiva que evalúa si la hora es posterior a las 14:30
# El argumento es la hora que se quiere evaluar y su tipo de dato es string
# El valor de retorno de la función es un valor de tipo booleano (será True si la hora es posterior a las 14:30:00)
def calcular_tarde(h):
  if h > '14:30:00':
    return True

# Las regex que voy a necesitar
regex_hora = '\d{2}:\d{2}:\d{2}'
regex_direccion = '\S+@\S+'
regex_fecha = 'Date.+'

# Creo la lista vacía de lo que quiero imprimir
lista_correos = []

# Recorro la lista de líneas que me hice arriba
for x in lista_contiguas:
  # Encuentro el trozo de la hora
  lista_horas = re.findall(regex_hora, x)
  # Encuentro el trozo de la dirección
  lista_direcciones = re.findall(regex_direccion, x)
  # Recorro la lista de las horas
  for y in lista_horas:
    # Llamo a la función y la almaceno en la variable tarde
    tarde = calcular_tarde(y)
    # Si la variable tarde es True significa que el trozo de hora que entra en cada vuelta es posterior a las 14:30
    if tarde == True:
      # Si es True, la lista de correos se va extendiendo con los trozos de direcciones metidos en la lista
      lista_correos.extend(lista_direcciones)

print('A continuación se imprimen por pantalla las direcciones de correo que han enviado correos más tarde de las 14:30:\n')
# Recorro la lista de correos e imprimo cada elemento
for x in lista_correos:
  print(x)

# Ahora imprimo la lista de los correos en formato string separados por comas
print('\nLos correos son\n')
# Para eso hago un join en la lista de los correos y le pongo el separador coma
print(', '.join(lista_correos))

# Esto lo he hecho para que se vean las fechas al lado de cada dirección porque hay direcciones que han enviado varios correos tardíos
print('\nA continuación se imprimen por pantalla las direcciones de correo y la fecha completa, ya que algunas direcciones han enviado más de un correo tardío, que han enviado correos más tarde de las 14:30:\n')

# Creo la lista vacía
lista_definitiva = []

# Recorro la lista de líneas que me hice arriba
for x in lista_contiguas:
  # Encuentro el trozo de la hora
  lista_horas = re.findall(regex_hora, x)
  # Encuentro el trozo de la dirección
  lista_direcciones = re.findall(regex_direccion, x)
  # Encuentro el trozo de la fecha
  lista_fechas = re.findall(regex_fecha, x)
  # Recorro la lista de las horas
  for y in lista_horas:
    # Llamo a la función y la almaceno en la variable tarde
    tarde = calcular_tarde(y)
    # Si la variable tarde es True significa que el trozo de hora que entra en cada vuelta es posterior a las 14:30:00
    if tarde == True:
      # Si es True, se añaden a la lista definitiva las líneas contiguas
      lista_definitiva.append(x)
      # Imprimo la dirección y la fecha, una al lado de la otra, de cada correo tardío
      print(lista_direcciones, lista_fechas)

# Esto lo he hecho para que se vea más bonito
print('\nHe creado la lista definitiva en un formato más legible:\n')
# Recorro la lista definitiva en la que se han añadido las líneas enteras de los correos tardíos e imprimo cada elemento
for x in lista_definitiva:
  print(x, '\n')
```

```
print('Somos Correctus, el asistente del corrector de textos que le ayuda a escribirles cartas, de forma automática, a los clientes con los presupuestos solicitados.')
print('Díganos su nombre:')
# Almaceno lo que escribe el usuario en la variable nombre_corrector
nombre_corrector = input()
print('Hola,', nombre_corrector + '. Le vamos a hacer varias preguntas para ayudarle.')

# Creo una lista vacía para almacenar los nombres de los clientes
lista_nombres_clientes = []
print('Escriba en cada casilla el nombre de cada uno de sus clientes. Escriba "fin" cuando no quiera introducir ningún cliente más:')
while True:
  # Almaceno lo que escribe el usuario en la variable nombre_cliente
  nombre_cliente = input()
  # Este bucle solo se rompe con la palabras 'fin'
  if nombre_cliente == 'fin':
    break
  # Añado los nombre_cliente a la lista
  lista_nombres_clientes.append(nombre_cliente)

# Creo una lista vacía para almacenar los nombres de los textos
lista_nombres_textos = []
# Itero en la lista_nombres_clientes tantas veces como sea su longitud
for x in range(len(lista_nombres_clientes)):
  print('¿Cómo se llama el texto de su cliente', lista_nombres_clientes [x] + '?')
  # Almaceno lo que escribe el usuario en la variable nombre_texto
  nombre_texto = input()
  # Añado los nombre_texto a la lista
  lista_nombres_textos.append(nombre_texto)

# Función productiva que calcula el presupuesto
# Los argumentos son c (correccion) de tipo cadena, i (intervencion) de tipo cadena y m (matrices) de tipo entero
# El valor de retorno es el presupuesto y es de tipo flotante
def calcula_presupuesto(c, i, m):
  if 0 < m < 10000:
    # Asigno un presupuesto mínimo si el número de matrices es muy bajo
    presupuesto = 30
  # Claculo el presupuesto en función del tipo de corrección, del nivel de intervención y del número de matrices
  else:
    if c == 'ortotipográfica':
      if i == 'alta':
        presupuesto = (m/1000) * 1.6
      if i == 'media':
        presupuesto = (m/1000) * 1.4
      if i == 'baja':
        presupuesto = (m/1000) * 1.2
    if c == 'estilo':
      if i == 'alta':
        presupuesto = (m/1000) * 2.2
      if i == 'media':
        presupuesto = (m/1000) * 2
      if i == 'baja':
        presupuesto = (m/1000) * 1.8
    if c == 'ambas':
      if i == 'alta':
        presupuesto = (m/1000) * 2.5
      if i == 'media':
        presupuesto = (m/1000) * 2.4
      if i == 'baja':
        presupuesto = (m/1000) * 2.3 
  return presupuesto

# Creo listas vacías para almacenar los datos necesarios para calcular el presupuesto
lista_correcciones = []
lista_intervenciones = []
lista_matrices = []
# Creo una lista vacía para almacenar los presupuestos calculados
lista_presupuestos = []
print('Ahora vamos a calcular el presupuesto.')
# Itero en la lista_nombres_textos tantas veces como sea su longitud
for x in range(len(lista_nombres_textos)):
  print('¿Qué tipo de corrección le va a hacer al texto', lista_nombres_textos [x] + '? Escriba "ortotipográfica", "estilo" o "ambas":')
  # Almaceno lo que escribe el usuario en la variable correccion
  correccion = input()
  # Aquí obligo a que las respuestas válidas lo sean aunque se escriban en mayúsculas
  correccion = correccion.lower()
  # Aquí pongo un while para evitar que el programa se pare ante una respuesta no deseada
  while correccion.lower() != 'ortotipográfica' and correccion.lower() != 'estilo' and correccion.lower() != 'ambas':
    correccion = input('Escriba, por favor, si la corrección es "ortotipográfica", "estilo" o "ambas": ')
    # Aquí obligo a que las respuestas válidas lo sean aunque se escriban en mayúsculas
    correccion = correccion.lower()
  # Añado las correccion a la lista
  lista_correcciones.append(correccion)
  print('¿Qué nivel de intervención necesita el texto', lista_nombres_textos [x] + '? Escriba "alta", "media" o "baja":')
  # Almaceno lo que escribe el usuario en la variable intervencion
  intervencion = input()
  # Aquí obligo a que las respuestas válidas lo sean aunque se escriban en mayúsculas
  intervencion = intervencion.lower()
  # Aquí pongo un while para evitar que el programa se pare ante una respuesta no deseada
  while intervencion.lower() != 'alta' and intervencion.lower() != 'media' and intervencion.lower() != 'baja':
    intervencion = input('Escriba, por favor, si la intervención necesaria es "alta", "media" o "baja": ')
    # Aquí obligo a que las respuestas válidas lo sean aunque se escriban en mayúsculas
    intervencion = intervencion.lower()
  # Añado los intervencion a la lista
  lista_intervenciones.append(intervencion)
  # Creo una variable de control de tipo booleano para el bucle
  control = True
  while control:
    try:
      print('¿Cuántas matrices tiene el texto', lista_nombres_textos [x] + '?')
      # Almaceno lo que escribe el usuario en la variable matrices y lo paso a entero
      # Esta línea puede dar error si el usuario no mete un entero
      matrices = int(input())
      # Si el número de matrices es <= 0, no tiene sentido, se lo digo al usuario y continúa el bucle
      if matrices <= 0:
        print('El número de matrices a corregir debe ser mayor que 0')
        continue
      # Añado los matrices a la lista
      lista_matrices.append(matrices)
      # El bucle se rompe cuando cambio la variable de control
      control = False
    except:
      # Si la línea del try ha dado error, se imprime lo siguiente y continúa el bucle
      print('Escriba el número de matrices con dígitos, por favor:')
      continue
  # Llamo a la función calcula_presupuesto asignando su salida a la variable presupuesto
  presupuesto = calcula_presupuesto(correccion, intervencion, matrices)
  # Añado los presupuesto a la lista
  lista_presupuestos.append(presupuesto)

# Recorro la lista_correcciones
for index, value in enumerate(lista_correcciones):
  # Si el valor es 'estilo' lo cambio por 'de estilo' para que tenga sentido en la impresión de las cartas
  if value == 'estilo':
    lista_correcciones[index] = 'de estilo'
  # Si el valor es 'ambas' lo cambio por 'ortotipográfica y de estilo' para que tenga sentido en la impresión de las cartas
  if value == 'ambas':
    lista_correcciones[index] = 'ortotipográfica y de estilo'

print('Sus cartas son:')
# Recorro las listas tantas veces como sea la longitud de lista_nombres_clientes
for x in range(len(lista_nombres_clientes)):
  print('Buenos días, ' + lista_nombres_clientes [x] + ':')
  print('En relación con el texto ' + lista_nombres_textos [x] + ' que me envió de ' + str(lista_matrices [x]) + ' matrices para que valorara,')
  print('el presupuesto para una corrección ' + lista_correcciones [x] + ' es de ' + str(lista_presupuestos [x]) + ' euros.')
  print('No dude en ponerse en contacto conmigo para cualquier aclaración o consulta.')
  print('Reciba un cordial saludo.\n' + nombre_corrector)

# Creo una lista vacía para almacenar los strings de las cartas
lista_cartas = []
# Recorro las listas tantas veces como sea la longitud de lista_nombres_clientes
for x in range(len(lista_nombres_clientes)):
  # La variable carta es la concatenación de los string de las listas
  carta = 'Buenos días, ' + lista_nombres_clientes [x] + ':\nEn relación con el texto ' + lista_nombres_textos [x] + ' que me envió de ' + str(lista_matrices [x]) + ' matrices para que valorara,\nel presupuesto para una corrección ' + lista_correcciones [x] + ' es de ' + str(lista_presupuestos [x]) + ' euros.\nNo dude en ponerse en contacto conmigo para cualquier aclaración o consulta.\nReciba un cordial saludo.\n' + nombre_corrector
  # Añado los carta a la lista
  lista_cartas.append(carta)

# Creo un fichero de escritura
manejador_escritura = open('fichero_escritura.txt', 'w')
intro = nombre_corrector + ', Estas son las cartas que hemos redactado con la información facilitada:\n'
# Escribo la entradilla en el fichero
manejador_escritura.write(intro)
# Convierto la lista_cartas en un string separando los elementos de la lista con vueltas de carro
carta = ('\n'.join(lista_cartas))
# Escribo esta cadena en el fichero
manejador_escritura.write(carta)
# Cierro el fichero
manejador_escritura.close()
print('Hemos generado un archivo llamado "fichero_escritura.txt", que encontrará a la izquierda, con las cartas.')

print('Además, ofrecemos un servicio de tratamiento de textos.')
print('Le hacemos una pequeña demostración de algunas de las utilidades, pero recuerde que puede ponerse en contacto con nosotros para que ajustemos el programa a sus necesidades.')

texto_demo = '''ENAIRE, a través del Departamento de Coordinación Operativa del Espacio Aéreo (COOP), ha sido el responsable de coordinar los vuelos sobre el espacio aéreo español de SOFIA (Stratospheric Observatory for Infrared Astronomy), el observatorio aéreo más grande del mundo, que ha sido desarrollado y operado por la NASA.
El objetivo de SOFIA es estudiar el universo, en un espectro infrarrojo, para entender el desarrollo de las galaxias y la formación y evolución tanto de las estrellas como de los sistemas planetarios.
SOFIA tiene previsto efectuar un total de veinte vuelos científicos por toda Europa hasta el próximo 12 de marzo, el primero de los cuales despegó el pasado 9 de febrero. Algunos de estos vuelos se producen sobre el espacio aéreo español y son gestionados por los controladores aéreos de ENAIRE, que se encargan de facilitar en tiempo real la labor de investigación de SOFIA. 
Para desarrollar esta misión se ha acoplado un telescopio a bordo de un Boeing 747 que, desde la estratosfera, recoge informaciones sin interferencias atmosféricas, algo imposible de lograr en los telescospios tradicionales basados en la tierra. 
Entre los fenómenos astronómicos que estudia SOFIA destacan el nacimiento y muerte de las estrellas, la formación de nuevos sistemas solares, la identificación de moléculas espaciales complejas y los agujeros negros, así como planetas, cometas y asteroides de nuestro sistema solar.'''

# Función productiva que cuenta las matrices del texto
# Su argumento es el texto y es de tipo cadena
# Su salida es el contador y es de tipo entero
def contar_matrices(texto):
  # Creo un contador a 0 y le añado 1 cada vez que encuentre en el texto un caracter o un espacio
  contador = 0
  while texto[contador:]:
    contador = contador + 1
  return contador

import re

# Función productiva que busca las palabras con tilde del texto
# Su argumento es el texto y es de tipo cadena
# Su salida es la lista de palabras tildadas y es de tipo lista
def buscar_tildes(texto):
  # Creo la regex que busco
  regex_pp_tildadas = '\w*[áéíóú]\w*'
  # Creo la lista_pp_tildadas con el método findall
  lista_pp_tildadas = re.findall(regex_pp_tildadas, texto)
  return lista_pp_tildadas

import re, string

# Función productiva que cuenta las palabras del texto
# Su argumento es el texto y es de tipo cadena
# Su salida es la longitud de la lista de palabras y es de tipo entero
def contar_palabras(texto):
  # Le quito la puntuación al texto por si hubiera un caracter suelto
  texto_sin_puntuacion = re.sub('[%s]' % re.escape(string.punctuation), '', texto)
  # Creo una lista_palabras separando el texto por espacios
  lista_palabras = texto_sin_puntuacion.split()
  numero_palabras = len(lista_palabras)
  return numero_palabras

# Función productiva que cuenta la longitud de las palabras del texto
# Su argumento es el texto y es de tipo cadena
# Su salida es cada palabra con su longitud y es de tipo diccionario
def quitar_punt_y_contar_long(texto):
  # Le quito la puntuación al texto para que no cuente los caracteres pegados a las letras como parte de la palabra
  texto_sin_puntuacion = re.sub('[%s]' % re.escape(string.punctuation), '', texto)
  # Creo una lista_palabras separando el texto por espacios
  lista_palabras = texto_sin_puntuacion.split()
  # Creo un diccionario vacío
  diccionario = {}
  # Recorro la lista_palabras y almaceno cada palabra como clave y su longitud como valor
  for x in lista_palabras:
    diccionario[x] = len(x)
  return diccionario

# Llamo a la función contar_matrices asignando su salida a la variable contador
contador = contar_matrices(texto_demo)
print('El texto tiene', contador, 'matrices.')
# Llamo a la función buscar_tildes asignando su salida a la variable lista_pp_tildadas
lista_pp_tildadas = buscar_tildes(texto_demo)
print('El texto contiene las siguientes palabras con tilde:\n', lista_pp_tildadas)
# Llamo a la función contar_palabras asignando su salida a la variable numero_palabras
numero_palabras = contar_palabras(texto_demo)
print('El texto tiene', numero_palabras, 'palabras.')
# Llamo a la función quitar_punt_y_contar_long asignando su salida a la variable diccionario
diccionario = quitar_punt_y_contar_long(texto_demo)
print('Le mostramos la longitud de cada palabra del texto:\n', diccionario)

print('Ahora puede probar usted pegando el texto que quiera en la siguiente casilla:')
while True:
  print('Escriba un texto (o "salir" para finalizar el programa), por favor: ')
  # Almaceno lo que escribe el usuario en la variable texto
  texto = input()
  # Este bucle solo se rompe con la palabra 'salir' e imprimo la despedida
  if texto == 'salir':
    print('Esperamos haberle servido de ayuda. Contacte con nosotros para personalizar los servicios, se sorprenderá. Gracias y hasta la próxima.')
    break
  # Llamo a la función contar_matrices asignando su salida a la variable contador
  contador = contar_matrices(texto)
  print('El texto tiene', contador, 'matrices.')
  # Llamo a la función buscar_tildes asignando su salida a la variable lista_pp_tildadas
  lista_pp_tildadas = buscar_tildes(texto)
  print('El texto contiene las siguientes palabras con tilde:\n', lista_pp_tildadas)
  # Llamo a la función contar_palabras asignando su salida a la variable numero_palabras
  numero_palabras = contar_palabras(texto)
  print('El texto tiene', numero_palabras, 'palabras.')
  # Llamo a la función quitar_punt_y_contar_long asignando su salida a la variable diccionario
  diccionario = quitar_punt_y_contar_long(texto)
  print('Le mostramos la longitud de cada palabra del texto:\n', diccionario)
```
